```python
def transposed_matrix(matrix):
    transposed=([matrix[j][i] for j in range(len(matrix))] for i in range(len(matrix[0])))
    return transposed
matrix=[
    [1,2,3],
    [4,5,6],
    [7,8,9]
]
transposed=transposed_matrix(matrix)
print("original matrix:")
for i in matrix:
    print(i)
print("transposed matrix:")
for r in transposed:
    print(r)
```

```python
def flat(nested):
    flat_lst=[j for i in nested for j in i]
    return flat_lst
nested=[
    [1,2,3],
    [4,5,6],
    [7,8,9]
]
flat_lst=flat(nested)
print("original matrix:")
print(nested)
print("flattend matrix:")
print(flat_lst)
```

```python
def reversed(lst):
   return lst[::-1]

lst=[1,2,3,4,5]
reverse=reversed(lst)
print("original lst:")
print(lst)
print("reversed list:")
print(reverse)
```

```python
def rotated(lst,n):
    return lst[n:]+lst[:n]
lst=[1,2,3,4,5]
rotate=rotated(lst,3)
print("original:")
print(lst)
print("rotated list:")
print(rotate)
```

```python
def intersection(lst1,lst2):
    intersect=[i for i in lst1 if i in lst2 ]
    return intersect
lst1=[1,2,3,5,6,7]
lst2=[1,2,5,7,8,9]
intersect=intersection(lst1,lst2)
print(lst1)
print(lst2)
print(intersect)
```
