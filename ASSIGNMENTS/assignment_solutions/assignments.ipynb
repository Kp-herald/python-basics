```python
# assignment 12
n= int(input("enter a number"))
fact =1
i=1
while i<=n:
    fact*=i
    i+=1
print(f"factorial of {n} is {fact}")
```

```python
# assignment 13
n=int(input("enter a number"))
sum=0
while n>0:
    digit=n%10
    sum+=digit
    n//=10

print(sum)
```

```python
# assignment 14
num=int(input("enter a number"))
if num>1:
    for i in range(2,num):
        if num%i==0:
            print ("not a prime")
            break
    else:
        print("it is a prime")
else:
    print("not a prime")
```

```python
# assignment 15
n=int(input("enter a number for fibonacci series"))
a,b=0,1
count=0
while count<n:
    print(a,end=" ")
    a,b=b,a+b
    count+=1

```
