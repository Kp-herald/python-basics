```python
def mini(tpl):
    return min(tpl)
def maxi(tpl):
    return max(tpl)
def sumof(tpl):
    return sum(tpl)

tpl=(1,2,3,4,0,5)
print(mini(tpl))
print(maxi(tpl))
print(sumof(tpl))
```

```python
def min_in_tuple(tpl):
    return min(tpl)

def max_in_tuple(tpl):
    return max(tpl)

def sum_of_tuple(tpl):
    return sum(tpl)

sample_tpl = (1, 2, 3, 4, 5)
print(f"Minimum: {min(sample_tpl)}")
print(f"Maximum: {max_in_tuple(sample_tpl)}")
print(f"Sum: {sum_of_tuple(sample_tpl)}")
```
