```python
# assignment 9
import numpy as np
data_type=[('name','U10'),('age','i4'),('weight','f4')]
data=np.array([('herald',22,57.3),('jesus',33,109),('pramod',26,61.3)],dtype=data_type)
print(f"original:\n{data}")

sorted=np.sort(data,order="name")
print(f"sorted\n{sorted}")
```

```python
import numpy as np

data_type = [('x', 'i4'), ('y', 'i4')]
data = np.array([(1, 2), (3, 4), (5, 6)], dtype=data_type)

points = np.column_stack((data['x'], data['y']))

distances = np.linalg.norm(points[:, np.newaxis] - points, axis=2)

print(distances)
```
