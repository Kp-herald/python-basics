### Organizing students grade

```python
## grades of students
grades=[90,89,78,86,97]
# adding grade
grades.append(98)
## average of grades
average=sum(grades)/len(grades)
print(f"average of grades :{average:.2f}")
## max and min of grades
highest=max(grades)
lowest=min(grades)
print(f"highest grade is :{highest}")
print(f"lowest grade is :{lowest}")
```

```python
## feedback
feedback=["good service","excellent work","satisfied with work","could be better"]
## adding another feedback
feedback.append("not happy with service")
## counting positive feedbacks
p_feedback=sum(1 for i in feedback if "great" in i.lower() or "excellent" in i.lower())
print(f"positive feedback coount:{p_feedback}")
# entering all other feedbackz
print("all feedbacks:")
for item in feedback:
    
    print(f"-{item}")
```
