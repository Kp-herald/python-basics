```python
def reading(path):
    with open(path,"r") as file:
        lines=file.readlines()
        length=len(lines)
        w_count=sum(len(line.split()) for line in lines)
        c_count=sum(len(line) for line in lines)
        return length,w_count,c_count
file="sample.txt"
line,wcount,ccount=reading(file)
print(f"line count:{line}\n word count:{wcount} \n char count:{ccount}")
```
