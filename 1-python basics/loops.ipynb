```python
#while loop

count=2
while count<=10:
    print(count)
    count=count+1
```

```python
#break
for i in range(10):
    if i==4:
        break
    print(i)
```

```python
#continue
for i in range(10):
    if i==4:
        continue
    print(i)
```

```python
#continue
for i in range(10):
    if i%2!=0:
        continue
    print(i)
```

```python
#pass
for i in range(5):
    if i==4:
        print("checking number but doing nothing")
        pass
    print(i)
```

```python
# nested loops
for i in range(3):
    for j in range(3):
        print (f"i:{i} and j:{j}")
```

```python
# adding n natural numbers by using while and for loops 
n=10
sum=0
count=1
while count<=n:
    sum=sum+count
    count=count+1
print(f"the sum of {n} natural numbers is: {sum}")
```

```python
# for loop
sum=0
for i in range(11):
    sum=sum+i
print(f"sum of first n natural numbers is :{sum}")
```

```python
#prime number between 1 - 100
for num in range(1,101):
    if num>1:
        for i in range(2,num):
            if num%i==0:
                break
        else:
            print(num)
        
```
