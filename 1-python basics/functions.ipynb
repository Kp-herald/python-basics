```python
def even_or_odd(num):
    if num%2==0:
        print(f"{num} is even")
    else:
        print(f"{num} is odd")
```

```python
# positional arguments
def num(*args): #we can write *herald,*something
    for val in args:
        print(f"{val}")
num(1,2,3,4,5,6.8,"herald","jesus")
```

```python
#keyword arguments
def pair(**kwargs):
    for i,j in kwargs.items():
        print(f"{i}:{j}")
pair(name="herald",age=22,purpose="to live with god")
```

```python
#both keywords and positional arguments
def both(*args,**kwargs):
    for i in args:
        print(f"{i}")
    for key,val in kwargs.items():
        print(f"{key}:{val}")
    
both(1,2,2,4,4,5,56,"herald",name="pramod",age=22)
```

```python
#converting celsius to farehheit and vice versa
def convert(temp,unit):
    if unit=='C':
        return temp * 9/5 + 32
    elif unit=="F":
        return (temp-32) * 5/9
    else:
        return None
print(convert(45,"C"))
print(convert(93,"F"))
print(convert(70,""))
```

```python
# password strength checker
def is_strong(pwd):
    if len(pwd)<8:
        return False
    if not any(char.islower() for char in pwd):
        return False
    if not any(char.isdigit() for char in pwd):
        return False
    if not any(char.isupper() for char in pwd):
        return False
    if not any(char in '@!#($|%&*^_-' for char in pwd):
        return False
    return True

# calling function
print(is_strong("rjah"))
print(is_strong("RajKing"))
print(is_strong("Jesus_L0ves"))
print(is_strong("Rinkukph@123"))
print(is_strong("rjah0eug3W"))
```

```python
# calculating the total cost of the items
def calc(cart):
    total=0
    for item in cart:
        total+=item['price'] * item['quantity']
    return total
product=[
    {"name":"snickers","price":20,"quantity":5},
    {"name":"chocos","price":10,"quantity":2},
    {"name":"chips","price":10,"quantity":6},
    {"name":"car","price":520,"quantity":1},
]
print(calc(product))
```

```python
# check is it is palindrome
def palindrome(val):
    val=str(val)
    return val==val[::-1]
print(palindrome("ramar"))
print(palindrome(12321))
print(palindrome(123231))
print(palindrome("herald"))
```

```python
#factorial of a number
def fact(num):
    if num==0:
        return 1
    else:
        num=num*fact(num-1)
    return num
print(fact(6))
    
```

```python
# read a file and count frequency of each word
def count(filepath):
    wordc={}
    with open(filepath,'r') as file:
        for line in file:
            words=line.split()
            for word in words:
                word=word.lower().strip('@$%&*!')
                wordc[word]=wordc.get(word,0)+1
            return wordc
print(count('sample.txt'))
                
```

```python
# email validation
import re
def valid(email):
    pattern=r'^[a-zA-Z0-9.+_-]+@[a-zA-Z0-9-]+\.[a-zA-Z]'
    return re.match(pattern,email) is not None
print(valid("kpherald7@gmail.com"))
print(valid("kpherald-"))
```
