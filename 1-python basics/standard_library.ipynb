```python
# data serialization

import json
data={"name":"jesus","type":"truth","follow":"him"}
print(type(data))

json_str=json.dumps(data)
print(json_str)
print(type(json_str))

json_con=json.loads(json_str)
print(json_con)
print(type(json_con))
```

```python
## csv 

import csv
with open("simple.txt",mode="w",newline="") as file :
    writer=csv.writer(file)
    writer.writerow(["name","age",'follows'])
    writer.writerow(["herald",22,"christ"])
    writer.writerow(["pramod",23,"jesus"])

with open("simple.txt",mode="r") as file:
    read=csv.reader(file)
    for i in read:
        print(i)
```
