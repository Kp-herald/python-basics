## Creating a to do list


```python
to_do_list=["pay the bills","learn data analyst","clean your room"]

## adding to do lists
to_do_list.append("tell tution")
to_do_list.append("pray for the job")

## removing to do list
to_do_list.remove("clean your room")

## checking if the task in to do list
if "pray for the job" in to_do_list:
    print("remainder: don't forget to pray for the job")
## remaining to do works
print("remaining tasks to complete")
for remain in to_do_list:
    print(f"-{remain}")
```
