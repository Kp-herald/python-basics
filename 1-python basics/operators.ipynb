```python
#operators
a=10
b=3
sum=a+b #addition
sub=a-b #substraction
mul=a*b #multiplication
div=a/b #division
floor_div=a//b #floor division
mod=a%b #modulus
expo=a**b #exponents

print(sum)
print(sub)
print(mul)
print(div)
print(floor_div)
print(mod)
print(expo)
```

### logical operators

```python
# logical operators are and,or,not
# AND -> if both are true then it is true otherwise false
# OR -> if both are false then it is false otherwise true
# NOT -> reverse of a boolean

#AND
a=True
b=True
print(a and b)

a=True
b=False
print(a and b)

a=False
b=True
print(a and b)

a=False
b=False
print(a and b)
```

```python
# OR
a=True
b=True
print(a or b)

a=True
b=False
print(a or b)

a=False
b=True
print(a or b)

a=False
b=False
print(a or b)
```

```python
# not
a=True
print(not a)

a=False
print(not a)
```
